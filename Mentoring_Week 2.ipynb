# Task 1. Calculate Average Sessions

**Task Descriptions**

- PacBook is a Social Media platform and they want to know what is their user Average Sessions and Session Duration for each data
- Given a three input for `load_time` (in minutes) and `exit_time` (in minutes)

- To calculate Session Duration, you can use this formula

$$
\text{Session}_\text{duration} = \text{exit}_\text{time} - \text{load}_\text{time}
$$

- To calculate Average Sessions, you can use this formula

$$
\text{Average}_\text{sessions} = \frac{\sum_{n=1}^{3}{\text{Session_duration}_{n}}}{\text{N_SESSIONS}}
$$

- But there are some constraints in this task
  - If the input for `load_time` and `exit_time` is less than zero or negative values, it will print the result **`"Not a Valid Input"`**

**Input - 1**

```py
load_time_1 = 5
exit_time_1 = 10

load_time_2 = 2
exit_time_2 = 15.5

load_time_3 = 1
exit_time_3 = 22.3
```

**Output - 1**

```py
Session 1 Duration (in minutes): 5
Session 2 Duration (in minutes): 13.5
Session 3 Duration (in minutes): 21.3
Average Session Duration (in minutes): 13.27
```

**Input - 2**

```py

```

**Output - 2**



  - Celcius to Fahrenheit, vice versa

````python
# Write your code here
temp = 20
from_unit = 'Celcius'
to_unit = 'Fahrenheit'

#kondisi jika from_unit adalah 'Celcius'
if from_unit == 'Celcius':

  #kondisi jika to_unit adalah 'Fahrenheit'
  #melakukan konversi celcius to fahrenheit dan print hasil konversi-nya
  if to_unit == 'Fahrenheit':
    convert = (temp * 9/5) + 32
    convert = round(convert,2)
    print(f'{temp} {from_unit} is equal to {convert} {to_unit}.')

  #kondisi jika to_unit adalah 'Kelvin'
  #melakukan konversi celcius to kelvin dan print hasil konversi-nya
  elif to_unit == 'Kelvin':
    convert = temp + 273.15
    convert = round(convert,2)
    print(f'{temp} {from_unit} is equal to {convert} {to_unit}.')

#kondisi jika from_unit adalah 'Fahrenheit'
elif from_unit == 'Fahrenheit':

  #kondisi jika to_unit adalah 'Celcius'
  #melakukan konversi fahrenheit to Celcius dan print hasil konversi-nya
  if to_unit == 'Celcius':
    convert = (temp - 32) * 5/9
    convert = round(convert,2)
    print(f'{temp} {from_unit} is equal to {convert} {to_unit}.')

  #kondisi jika to_unit adalah 'Kelvin'
  #melakukan konversi fahrenheit to kelvin dan print hasil konversi-nya
  elif to_unit == 'Kelvin':
    convert = (temp - 32) * 5/9 + 273.15
    convert = round(convert,2)
    print(f'{temp} {from_unit} is equal to {convert} {to_unit}.')

#kondisi jika from_unit adalah 'Kelvin'
elif from_unit == 'Kelvin':

  #kondisi jika to_unit adalah 'Celcius'
  #melakukan konversi kelvin to Celcius dan print hasil konversi-nya
  if to_unit == 'Celcius':
    convert = temp - 273.15
    convert = round(convert,2)

    print(f'{temp} {from_unit} is equal to {convert} {to_unit}.')

  #kondisi jika to_unit adalah 'Fahrenheit'
  #melakukan konversi kelvin to fahrenheit dan print hasil konversi-nya
  elif to_unit == 'Fahrenheit':
    convert = (temp - 273.15) * 9/5 + 32
    convert = round(convert,2)
    print(f'{temp} {from_unit} is equal to {convert} {to_unit}.')

#kondisi else jika nilai from_unit dan to_unit tidak sesuai
else:
  print('Something wrong with your input!')
````

# Task 3. Extract NIK (Nomor Induk Kependudukan) Data

Category: Hard

---

**Task Descriptions**
- NIK is not some random numbers. You can extract information from the NIK.
  - The first six digits are a province, regency, and district code
  - The next six digits are the date of birth; DD-MM-YY, respectively
    - For men, the date of birth ranged between 01-31
    - For women, the date of birth is added with 40, hence ranged between 41-71.
  - The last 4 digits are a serial number of NIK issuance.


- Given **two** NIK, your tasks are
  1. Extract the date of birth of each NIK
  2. Extract their gender.
  3. Extract whether they have similar gender.
  4. Extract whether they have similar month of birth.

**Input**

```python
nik_1 = 1571011709860003
nik_2 = 2404075109910009
```

**Output**
```
Date of Birth
- NIK 1 : 17-09-86
- NIK 2 : 11-09-91

Gender
- NIK 1 : Male
- NIK 2 : Female

Do they have similar gender?
No

Do they have similar month of birth?
Yes, it is on September
```

````python
nik_1 = 1571011709860003
nik_2 = 2404075109910009

#Substring date of birth dari nik 1
nik_1 = str(nik_1) #convert nik_1 menjadi string supaya bisa di substring
tanggal_1 = int(nik_1[6:8]) #convert variable tanggal_1 menjadi integer supaya bisa di operasikan pada branching menentukan gender
bulan_1 = nik_1[8:10]
tahun_1 = nik_1[10:12]
gender_1 = 'Male' #set default gender as male


#Substring date of birth dari nik 2
nik_2 = str(nik_2) #convert nik_2 menjadi string supaya bisa di substring
tanggal_2 = int(nik_2[6:8]) #convert variable tanggal_2 menjadi integer supaya bisa di operasikan pada branching menentukan gender
bulan_2 = nik_2[8:10]
tahun_2 = nik_2[10:12]
gender_2 = 'Male' #set default gender as male



#Branching untuk menentukan gender jika tanggal lahirnya > 31
#maka nilai akan dikurang 40 untuk tanggal lahir dan dikategorikan sebagai wanita
if tanggal_1 > 31:
  tanggal_1 -= 40
  gender_1 = 'Female'

elif tanggal_2 > 31:
  tanggal_2 -= 40

  gender_2 = 'Female'

else:
  pass

#penggabungan tanggal lahir, bulan_lahir, dan tahun lahir menggunakan -
#disini tanggal_1 dan tanggal_2 di convert lagi menjadi string supaya bisa di concatenation dengan '-'
dob_nik_1 = str(tanggal_1) + '-' + bulan_1 + '-' + tahun_1
dob_nik_2 = str(tanggal_2) + '-' + bulan_2 + '-' + tahun_2

#Print date of birth dari nik1 dan nik2
print('Date of Birth:')
print(f'NIK 1 : {dob_nik_1}')
print(f'NIK 2 : {dob_nik_2}')

#Print gender berdasarkan tanggal lahir yang sudah ditentukan sebelumnya
print()
print('Gender:')
print(f'NIK 1 : {gender_1}')
print(f'NIK 2 : {gender_2}')

#Compare gender
print()
print('Do they have similar gender?')

#Branching untuk compare gender_1 dan gender_2
if gender_1 == gender_2:
  print('Yes')
else:
  print('No')

#Compare month of birth
print()
print('Do they have similar month of birth?')
#Store month in list so we can print based on bulan_1 or bulan_2
months = ['January','February','March','April'
          ,'May','June','July','August','September'
          ,'October','November','December']

if bulan_1 == bulan_2:
  bulan_1 = int(bulan_1) - 1
  print(f'Yes, it is on {months[bulan_1]}.')

else:
  print('No, they dont have a similar month of birth.')

````

# Task 4. Calculate Pajak Penghasilan (Simplified)

Category: Hard

---

**Task Descriptions**
- According to Pemotongan Pajak Penghasilan PPh Pasal 21 Tahun 2016, we can calculate our income tax as (we'll write in Bahasa Indonesia for avoiding missinterpretation)

- **Penghasilan tidak kena pajak (PTKP)**
  - Penghasilan Rp 54.000.000 setahun untuk wajib pajak.
  - Tambahan Rp 4.500.000 diberikan jika wajib pajak menikah.
  - Tambahan Rp 4.500.000 lagi jika punya 1 orang anak atau Rp 9.000.000 jika punya 2 orang anak. Apabila jumlah tanggungan lebih dari itu, maka tetap dihitung 3 orang tanggungan.

- **Tarif Pajak Penghasilan**
  - 5% untuk penghasilan kena pajak (PKP) sampai dengan 60 Juta
  - 15% untuk PKP di atas 60 juta sampai dengan 250 Juta
  - 25% untuk PKP di atas 250 juta sampai dengan 500 juta
  - 30% untuk PKP di atas 500 juta sampai dengan 5 miliar
  - 35% untuk PKP di atas 5 miliar

````python
# Write your code here
monthly_income = 4_750_000
marital_status = 'single'
n_dependents = 0

#initiate variable
wajib_pajak = 54_000_000
wajib_pajak_menikah = 0
pajak_tanggungan = 0

#menghitung yearly income
yearly_income = monthly_income * 12

#kondisi berdasarkan marital_status = married
if marital_status == 'married':

  #jika n_dependents <=3 maka pajak tanggungan dikali jml tanggungan
  #total PTKP K total dari pajak menikah, tanggungan, dan wajib pajak
  if n_dependents <= 3:
    wajib_pajak_menikah = 4_500_000
    pajak_tanggungan = 4_500_000 * n_dependents
    total_PTKP_K = wajib_pajak_menikah + pajak_tanggungan + wajib_pajak

  #jika n_dependents >3 maka pajak tanggungan dikali 3
  #total PTKP K total dari pajak menikah, tanggungan, dan wajib pajak
  elif n_dependents > 3:
    wajib_pajak_menikah = 4_500_000
    pajak_tanggungan = 4_500_000 * 3
    total_PTKP_K = wajib_pajak_menikah + pajak_tanggungan + wajib_pajak

#jika marital_status single
#tidak ada pajak menikah dan pajak tanggungan
elif marital_status == 'single':
  wajib_pajak_menikah = 0
  pajak_tanggungan = 0
  total_PTKP_K = wajib_pajak

#jika marital_status tidak sesuai dengan married dan single
else:
  print('Marital status tidak diketahui')

#menghitung total penghasilan kena pajak (PKP)
pkp = yearly_income - total_PTKP_K

#kondisi jika pkp masuk kedalam berbagai kategori nilai yang akan mempengaruhi
#nilai pajak yang akan diberikan
if pkp <= 60_000_000:
  percent = (5/100)
  pph = percent * pkp

elif 60_000_000 < pkp <= 250_000_000:
  percent = (15/100)
  pph = percent * pkp

elif 250_000_000 < pkp <= 500_000_000:
  percent = (25/100)
  pph = percent * pkp

elif 500_000_000 < pkp <= 5_000_000_000:
  percent = (30/100)
  pph = percent * pkp

elif pkp > 5_000_000_000:
  percent = (35/100)
  pph = percent * pkp

#print semua hasil perhitungan yang sudah diproses sebelumnya
#print yearly_income, total_PTKP_K, pkp dan pph
print('Penghasilan netto setahun: ')
print(f'Rp {yearly_income:,}')
print()

print(f'PTKP K/{n_dependents}: ')
print(f'-Rp {total_PTKP_K:,}')
print()

print('Total penghasilan kena pajak: ')
print(f'Rp {pkp:,}')
print()

print('PPh 21 terutang setahun: ')
print(f'Rp {pph:,}')
````
